```python
import cv2 as cv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
import os

CATEGORIES = ["Negative", "Positive"]
TRAINDIR = "/kaggle/input/surface-crack/train"
VALIDDIR = "/kaggle/input/surface-crack/valid"
TESTDIR = "/kaggle/input/surface-crack/test"
IMG_SIZE = 200

training_data = []
valid_data = []
test_data = []

def create_training_data():
    for category in CATEGORIES:
        path = os.path.join(TRAINDIR, category)
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                image = os.path.join(path, img)
                img_array = cv.imread(image, cv.IMREAD_GRAYSCALE)
                resize_img = cv.resize(img_array, (IMG_SIZE, IMG_SIZE))
                ###
#                 plt.imshow(resize_img, cmap = 'gray')
#                 plt.show()
#                 break
                ###
                training_data.append([[resize_img], [class_num]])
            except Exception as e:
                pass
#         break

def create_testing_data():
    for category in CATEGORIES:
        path = os.path.join(TESTDIR, category)
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                image = os.path.join(path, img)
                img_array = cv.imread(image, cv.IMREAD_GRAYSCALE)
                resize_img = cv.resize(img_array, (IMG_SIZE, IMG_SIZE))
                test_data.append([[resize_img], [class_num]])
            except Exception as e:
                pass


def create_validation_data():
    for category in CATEGORIES:
        path = os.path.join(VALIDDIR, category)
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                image = os.path.join(path, img)
                img_array = cv.imread(image, cv.IMREAD_GRAYSCALE)
                resize_img = cv.resize(img_array, (IMG_SIZE, IMG_SIZE))
                valid_data.append([[resize_img], [class_num]])
            except Exception as e:
                pass

create_training_data()
create_testing_data()
create_validation_data()
```
